```python
import pandas as pd
import numpy as np
```

```python
games = pd.read_json('data/5/games.json')
games.isnull().value_counts()
games.dropna(inplace=True)
games['release_date'] = games['release_date'].str.replace('Dic','Dec')
games['release_date'] = games['release_date'].str.replace('Abr','Aug')
games['release_date'] = games['release_date'].str.replace('Ene','Dec')

games['release_date'] = pd.to_datetime(games['release_date'])
```

```python
games.set_index('title', inplace=True)
```
