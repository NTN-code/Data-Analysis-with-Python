```python
writer = pd.ExcelWriter('out_result.xlsx')
```

```python
excel = pd.ExcelFile('data/4/products.xlsx')
```

```python
excel.sheet_names
```

```python
products = excel.parse(sheet_name='Products', index_col=[0])
desc = excel.parse(sheet_name='Descriptions', index_col=[0])
merc = excel.parse(sheet_name='Merchants', index_col=[0])
```

```python
merc.isnull().value_counts()
```

```python
with writer:
    products.to_excel(writer, sheet_name='Prod')
    desc.to_excel(writer, sheet_name='Desc')
    merc.to_excel(writer, sheet_name='Merc')
```
