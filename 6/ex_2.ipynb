```python
import numpy as np
import pandas as pd
```

```python
column_names = ['color', 'director_name', 'num_critic_for_reviews', 'duration',
                'gross', 'movie_title', 'num_user_for_reviews', 'country',
                'cotent_rating', 'budget', 'title_year', 'imdb_score', 'genre']
movies_df = pd.read_csv('data/2/movies.csv',
                       error_bad_lines=False,
                       sep='|',
                       skiprows=3,
                       names=column_names,
                        na_values=['?'],
                        thousands=',',
                       )
movies_df = movies_df.dropna()
movies_df['title_year'] = pd.to_numeric(movies_df['title_year']).astype(np.int16)
# movies_df['title_year'] = movies_df['title_year'].astype(np.int16)

movies_df.set_index('movie_title', inplace=True)
```
