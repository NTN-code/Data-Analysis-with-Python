# Introduce with Visualization

```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
plt.figure(figsize=(14, 6))
plt.title('First graph')
x = np.arange(10)
y = np.flip(np.arange(10))
plt.plot(x,x)
plt.plot(x,np.flip(x))
```

```python
plt.figure(figsize=(14,6))
plt.title('Second')

plt.subplot(2,2,1)
plt.bar(x,y,)
plt.xlabel('XXX')
plt.ylabel('YYY')
plt.legend(['X func'])

plt.subplot(2,2,2)
plt.plot(x,y)
plt.legend(['X func and Y func'])


plt.subplot(2,2,3)
plt.hist(x, weights=x)

plt.subplot(2,2,4)
plt.scatter(x,y, s=100, c='g')
```

```python
plot_obj = plt.subplots(nrows=2, ncols=2, figsize=(14,6))
fig, ((ax1,ax2),(ax3,ax4)) = plot_obj
plot_obj
ax1.plot(np.random.randint(50), c='yellow')
ax2.plot(np.random.randn(50), c='b', linestyle=':')
ax3.plot(np.random.rand(50), c='r', marker='o', linewidth=3.0)
ax4.plot(np.random.randn(50), c='b', linestyle='--')
print(np.random.randint(50))
print(np.random.randn(50))
print(np.random.rand(50))
print()
```

```python
plt.figure(figsize=(14,6))
plt.title('DD')

plt.subplot2grid((4,4), (0,0), colspan=4)
plt.subplot2grid((4,4), (1,0), colspan=3)
plt.subplot2grid((4,4), (1,3), rowspan=3)
plt.subplot2grid((4,4), (2,0), colspan=2)
plt.subplot2grid((4,4), (2,2), rowspan=2)
plt.subplot2grid((4,4), (3,0), colspan=1)
plt.subplot2grid((4,4), (3,1), )


```

```python
N = 50 
x = np.random.randn(N)
y = np.random.randn(N)
colors = np.random.randn(N)
area = np.pi * (20 *np.random.rand(N))**2
plt.figure(figsize=(14,6))
plt.scatter(x,y, s=area, c =colors, alpha=1, cmap='Spectral')
plt.colorbar()
plt.show()
```
